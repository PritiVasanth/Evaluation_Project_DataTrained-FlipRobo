#### We can only skew numercial data. Categorical data and target cannot be skewed. 

```python
import numpy as np
from scipy.stats import skew, boxcox

# Select the columns
columns = ['Longitude', 'Latitude', 'Average Cost for two', 'Aggregate rating', 'Votes']

# Calculate skewness for each column
original_skew = {col: skew(df_pr[col]) for col in columns}
print(f"Original skewness: {original_skew}")

# Function to shift and transform data
def shift_and_transform(column_data):
    shifted_data = column_data - column_data.min() + 1  # Shift to make positive
    return {
        'log': np.log(shifted_data),
        'sqrt': np.sqrt(shifted_data),
        'boxcox': boxcox(shifted_data)[0]
    }

# Apply transformations and calculate skewness for each column
transformed_skew = {}
for col in columns:
    transformations = shift_and_transform(df_pr[col])
    log_skew = skew(transformations['log'])
    sqrt_skew = skew(transformations['sqrt'])
    boxcox_skew = skew(transformations['boxcox'])
    transformed_skew[col] = {
        'log': log_skew,
        'sqrt': sqrt_skew,
        'boxcox': boxcox_skew
    }
    print(f"{col} - Log-transformed skewness: {log_skew}, Square root-transformed skewness: {sqrt_skew}, Box-Cox-transformed skewness: {boxcox_skew}")

# Apply the best transformation (closest skewness to 0.5) for each column
for col in columns:
    skews = transformed_skew[col]
    best_transformation = min(skews, key=lambda x: abs(skews[x] - 0.5))
    df_pr[col] = shift_and_transform(df_pr[col])[best_transformation]
    print(f"Applied {best_transformation} transformation to {col}")

# Verify the final skewness
final_skew = {col: skew(df_pr[col]) for col in columns}
print(f"Final skewness: {final_skew}")
```

```python
maxAccu=0
maxRS=0
for i in range(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x_scaler,y1,test_size=.20, random_state=i)
    lr=LinearRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    acc=r2_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print('Maximum r2 score is',maxAccu,'on Random_state',maxRS)
```
