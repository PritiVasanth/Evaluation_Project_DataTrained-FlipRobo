#### There are five different columns for Generation GW-Hours from 2013 to 2019. This is an important feature for coal and hydro power plants. Missing data in this category is expected, as oil and gas power plants operate intermittently, and some renewable power plants like wind and tidal operate seasonally. New power plants commissioned between 2013 and 2019 will have some data available.
#### We will use mean or median imputation due to the varying generation capacities and yearly runtime differences among power plants. 
#### Another feature of interest is Other_fuel1,Other_fuel2, Other_fuel3 which also has missing values. This feature is not relevant for renewable energy sources like solar, wind, and hydro. For power plants where the concept of other fuel is applicable, we can impute based on domain knowledge:

- Coal power plants: Oil or cogeneration
- Industrial oil power plants: Gas
- Industrial gas power plants: Oil
#### We cannot define an alternative fuel for nuclear power plants due to their specific design. For renewable energy source-based power plants, no alternate fuel is needed. We will impute these with "Not Applicable." When encoding these labels, 'Not Applicable' will simply be an additional category, not affecting the final result.

# Treating column 'generation_gwh_2019' as the datatype id object

```python
# Step 1: Identify non-numeric values
non_numeric = df[~df['generation_gwh_2019'].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
print(non_numeric['generation_gwh_2019'])
```

```python
# Step 2: Replace or remove non-numeric values
df['generation_gwh_2019'] = pd.to_numeric(df['generation_gwh_2019'], errors='coerce')
```

```python
# filling null values using their median values for generation_gwh 
df['commissioning_year']=df['commissioning_year'].fillna(df['commissioning_year'].median())
df['generation_gwh_2013']=df['generation_gwh_2013'].fillna(df['generation_gwh_2013'].median())
df['generation_gwh_2014']=df['generation_gwh_2014'].fillna(df['generation_gwh_2014'].median())
df['generation_gwh_2015']=df['generation_gwh_2015'].fillna(df['generation_gwh_2015'].median())
df['generation_gwh_2016']=df['generation_gwh_2016'].fillna(df['generation_gwh_2016'].median())
df['generation_gwh_2017']=df['generation_gwh_2017'].fillna(df['generation_gwh_2017'].median())
df['generation_gwh_2018']=df['generation_gwh_2018'].fillna(df['generation_gwh_2018'].median())
df['generation_gwh_2019']=df['generation_gwh_2019'].fillna(df['generation_gwh_2019'].median())
```

```python
df.info()
```

#### We will have to drop other_fuel2 and other_fuel3 as the missing data is very high. and we wwill treat other_fuel1 with Mode Imputation. 

```python
df = df.drop(['other_fuel2', 'other_fuel3'], axis=1)
df.head()
```

```python
# let's check the mode of other_fuel1
df['other_fuel1'].mode()    
```

```python
# Replacing the null value by its mode Oil
df['other_fuel1']=df['other_fuel1'].fillna(df['other_fuel1'].mode()[0])
```

```python
df.info()
```

```python
#Now lets Treat capacity_mw. As only 6 values are missing we can use median imputation to replace 
df['capacity_mw']=df['capacity_mw'].fillna(df['capacity_mw'].median())
```

#### Predicting  Missing Target Values for primary_fuel
- For more sophisticated imputation,we can use a machine learning model to predict the missing target values based on other features in your dataset.

```python
from sklearn.ensemble import RandomForestClassifier  # For categorical target primary_fuel
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Separate the rows with and without missing target values
train_df = df[df['primary_fuel'].notna()]
test_df = df[df['primary_fuel'].isna()]
```

```python
# Separate features and target
X_train = train_df.drop(columns=['primary_fuel'])
y_train = train_df['primary_fuel']
X_test = test_df.drop(columns=['primary_fuel'])
```

```python
# Identify categorical columns
categorical_cols = X_train.select_dtypes(include=['object']).columns
```

```python
# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
# Combine the preprocessor with the model into a pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier())
])
```

```python
model.fit(X_train, y_train)
```

```python
# Predict the missing values
predicted_values = model.predict(X_test)
df.loc[df['primary_fuel'].isna(), 'primary_fuel'] = predicted_values
```

```python
maxAccu=0
maxRS=0
for i in range(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x_scaler,y,test_size=.30, random_state=i)
    lr=LinearRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    acc=r2_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print('Maximum r2 score is',maxAccu,'on Random_state',maxRS)
```
